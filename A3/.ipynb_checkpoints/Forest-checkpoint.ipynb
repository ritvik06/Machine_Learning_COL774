```python
from xclib.data import data_utils
import numpy as np
import math
import time
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import GridSearchCV
import sklearn
```

```python
X_train = data_utils.read_sparse_file('train_x.txt', force_header = True)
file = open('train_y.txt', 'r')
lines = file.readlines()
Y_train = []
for row in lines:
    Y_train.append(int(row[0]))
X_train = X_train.toarray()
S_train = np.zeros((len(X_train), (len(X_train[0])+1)))
for i in range(0, len(Y_train)):
    S_train[i, 0:len(X_train[0])] = X_train[i]
    S_train[i][len(X_train[0])] = Y_train[i]
file.close()

X_test = data_utils.read_sparse_file('test_x.txt', force_header = True)
file = open('test_y.txt', 'r')
lines = file.readlines()
Y_test = []
for row in lines:
    Y_test.append(int(row[0]))
X_test = X_test.toarray()
file.close()

X_valid = data_utils.read_sparse_file('valid_x.txt', force_header = True)
file = open('valid_y.txt', 'r')
lines = file.readlines()
Y_valid = []
for row in lines:
    Y_valid.append(int(row[0]))
X_valid = X_valid.toarray()
file.close()
```

```python
rfc = RandomForestClassifier(oob_score=True, n_jobs=-1)
parameters = {'n_estimators':[50, 150, 250, 350, 450], 'max_features':[0.1, 0.3, 0.5, 0.7, 0.9, 1.0], 'min_samples_split':[2, 4, 6, 8, 10]}
```

```python
start = time.time()
def scorer(estimator, X, Y):
    print(estimator.n_estimators, estimator.max_features, estimator.min_samples_split, estimator.oob_score_, time.time()-start)
    return estimator.oob_score_
clf = GridSearchCV(estimator=rfc, param_grid=parameters, scoring=scorer, cv=2)
clf.fit(X_train, Y_train)
print(time.time()-start)
```

```python
print(clf.best_params_)
```

```python
estimator = clf.best_estimator_
```

```python
print(estimator.oob_score_, estimator.score(X_train, Y_train), estimator.score(X_test, Y_test), estimator.score(X_valid, Y_valid))
```

```python
# Vary n_estimators
Accuracy_test_n_estimators = []
Accuracy_valid_n_estimators = []
Accuracy_train_n_estimators = []
n_estimator = [50, 150, 250, 350, 450]
for i in range(len(n_estimators)):
    estimator = RandomForestClassifier(n_estimators=n_estimator[i], oob_score=True, min_samples_split=10, max_features=0.1, n_jobs=-1)
    estimator.fit(X_train, Y_train)
    Accuracy_test_n_estimators.append(estimator.score(X_test, Y_test))
    Accuracy_valid_n_estimators.append(estimator.score(X_valid, Y_valid))
    Accuracy_train_n_estimators.append(estimator.score(X_train, Y_train))
    print(n_estimator[i], estimator.oob_score_, Accuracy_test_n_estimators[i], Accuracy_valid_n_estimators[i], Accuracy_train_n_estimators[i])
```

```python
# Vary max_features
Accuracy_test_features = []
Accuracy_valid_features = []
Accuracy_train_features = []
features = [0.1, 0.3, 0.5, 0.7, 0.9, 1.0]
for i in range(len(features)):
    estimator = RandomForestClassifier(n_estimators=350, oob_score=True, min_samples_split=10, max_features=features[i], n_jobs=-1)
    estimator.fit(X_train, Y_train)
    Accuracy_test_features.append(estimator.score(X_test, Y_test))
    Accuracy_valid_features.append(estimator.score(X_valid, Y_valid))
    Accuracy_train_features.append(estimator.score(X_train, Y_train))
    print(features[i], estimator.oob_score_, Accuracy_test_features[i], Accuracy_valid_features[i], Accuracy_train_features[i])
```

```python
# Vary min_samples_split
Accuracy_test_split = []
Accuracy_valid_split = []
Accuracy_train_split = []
split = [2, 4, 6, 8, 10]
for i in range(len(split)):
    estimator = RandomForestClassifier(n_estimators=350, oob_score=True, min_samples_split=split[i], max_features=0.1, n_jobs=-1)
    estimator.fit(X_train, Y_train)
    Accuracy_test_split.append(estimator.score(X_test, Y_test))
    Accuracy_valid_split.append(estimator.score(X_valid, Y_valid))
    Accuracy_train_split.append(estimator.score(X_train, Y_train))
    print(split[i], estimator.oob_score_, Accuracy_test_split[i], Accuracy_valid_split[i], Accuracy_train_split[i])
```
