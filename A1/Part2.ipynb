```python
import matplotlib.pyplot as plt
import numpy as np
import csv
import math
```

```python
theta = [[3],[1],[2]]
M = 10**6
s = 3 + np.random.normal(3,2,M) + 2*np.random.normal(-1,2,M) + np.random.normal(0,math.sqrt(2),M)
count, bins, ignored = plt.hist(s, 20, normed=True)
# Plot the distribution curve
plt.plot(bins, 1/(math.sqrt(14) * np.sqrt(2 * np.pi)) *
    np.exp(-(bins - 4)**2/(2 * 14)),linewidth=3, color='y')
plt.show()
```
